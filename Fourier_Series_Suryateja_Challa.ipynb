# Importing libraries

```python
import math
import sympy as sym
from sympy import pi
from matplotlib import pyplot as plt
```

### *t* is independent variable

```python
t = sym.symbols('t')
```

## **Problem 1**
### ***x(t) = -1, t<0;   1, t>=0***



fs2.png

## **Defining the piecewise function**

```python
x = sym.Piecewise((-1, t < 0 ), (1, t > 0))

ser = sym.fourier_series(x, (t, -pi, pi))
```

## **Printing upto five terms**

```python
ser.truncate(5)
```

## **Plot for different *n* values**

### n = 10

```python
sym.plot(ser.truncate(10), (t, -2*pi, 2*pi));
```

## n = 50

```python
sym.plot(ser.truncate(50), (t, -2*pi, 2*pi));
```

### n = 100

```python
sym.plot(ser.truncate(100), (t, -2*pi, 2*pi));
```

### n = 250

```python
sym.plot(ser.truncate(250), (t, -2*pi, 2*pi));
```

### n = 460 
#### *(Python cannot perform summation beyond this value)*

```python
sym.plot(ser.truncate(460), (t, -2*pi, 2*pi));
```

## **Problem 2**
### ***y(t)*** 
###         ***= -t-2, t>=2 and t<=-1;***   
###         ***= 1, t>=0    and t<=1 ;***
###         ***= -t+2, t>=1 and t<=2***

## **Defining piecewise function**

```python
y = sym.Piecewise( (-t-2, ((t >= -2) & (t <= -1))),
                   (t,     ((t >= -1) & (t <= 1))),
                   (-t+2, ((t >=  1) & (t <= 2))) )
```

## **Using the *'fourier_series'* function**

```python
ser1 = sym.fourier_series(y, (t, -2, 2))
```

## **Printing upto first five terms**

```python
ser1.truncate(5)
```

## **Plot for diiferent *'n'* values**

## n = 5

```python
sym.plot(ser1.truncate(5), (t, -2, 2));
```
